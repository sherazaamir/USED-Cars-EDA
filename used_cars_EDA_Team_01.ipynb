# **Introduction**: <a id="intnroduction"></a>
Exploratory Data Analysis (EDA) for used car analysis is like taking a closer look at all the information we have about secondhand vehicles. We examine things like how old the cars are, how much they've been driven, their prices, and other details. EDA helps us find patterns and insights in the data, making it easier to understand what factors affect the value of a used car. It's like opening the hood of a car to see what's inside before making a decision about buying or selling it.

### **About Dataset**
The provided contains various details and attributes associated with used cars. The target variable, which is the central focus of analysis, is the price of the used cars, and it is measured in lakhs.

**Source:** The dataset has been sourced from "Cars24."

**Data Type:** The data in this dataset is tabular, with rows and columns, where each row represents a specific used car listing, and each column represents a particular attribute or feature of these cars.

**Features:** 
- `Name`                -   Make and model of the car
- `Location`            -   Location or city of sale
- `Year`                -   Year of manufacture
- `Mileage`             -   Mileage
- `Kilometers_Driven`   -   Odometer (kilometers driven)

- `Transmission`        -   Transmission type (manual or automatic)
- `Owner_Type`          -   Number of owners
- `Engine`              -   Engine displacement
- `Power`               -   Engine horsepower
- `Seats`               -   Number of seats
- `New_Price`           -   Price when the car was new
- `Price`               -   Current Selling Price of the car

### Hypothesis-01: 
### Is there significant difference in preference of any feature `Fuel_Type` or `Mileage` ?

### Testing with 95% Confidence Interval. 

`H_0:` There is no preference difference in `Fuel_Type` or `Mileage` feature while buying/selling a car.

`H_1:` There is significant difference in preferring `Fuel_Type` or `Mileage` feature while buying/selling a car.

```python
# Create a contigency table of 'Mileage' and 'Fuel_Type' columns.

contigency_table = pd.crosstab(df['Mileage'], df['Fuel_Type'])
```

```python
# Perform chi-square test on the contigency table.

chi2, p_value, dof, expected = stats.chi2_contingency(contigency_table)
```

```python
# print the results by using the if else conditions.

print('p_value = ',p_value)
if p_value > 0.05:
    print("The preference in 'mileage' and 'fuel_type' categories are approximately equal.")
else:
    print("The preference in 'mileage' and 'fuel_type' categories are not equal.")
```

### **One-Sample T-test**

### `H_0:` The Sample Mean is equal to the Population Mean.
### `H_1:` The Sample Mean is not equal to the Population Mean.

```python
df.info()
```

##### Take a sample of 500 observations from the `Kilometers_Driven` column.

```python
# Specify the number of observations you want in the sample (e.g., 500)
sample_size = 500

# Take a random sample
random_sample = df['Kilometers_Driven'].sample(n=sample_size, random_state=42)
```

```python
# Calculate the mean of the population.
pop_mean = df['Kilometers_Driven'].mean()

# Performing the one-sample t-test
t_statistic, p_value = stats.ttest_1samp(random_sample, pop_mean)

# Print the results
print("One-sample t-test results:\n")
print("t-statistic:", t_statistic , "\n")
print("p-value:", p_value , "\n")

# Check if the p-value is less than 0.05
if p_value < 0.05:
    print("The sample mean is significantly different from the population mean.")
else:
    print("The sample mean is not significantly different from the population mean.")
```

### **Two-Sample T-test (independent)**

### `H_0:` The Sample_1 Mean is equal to the Sample_2 Mean.
### `H_1:` The Sample_1 Mean is not equal to the Sample_2 Mean.

```python
# Specify the number of observations you want in the sample (e.g., 500)
sample_size = 500

# Take a random sample
sample_1 = df['Kilometers_Driven'].sample(n=sample_size, random_state=42)
sample_2 = df['Kilometers_Driven'].sample(n=sample_size, random_state=42)

# Perform the two-sample t-test
t_statistic, p_value = stats.ttest_ind(sample_1, sample_2)

# Print the results
print("Two-sample t-test results:\n")
print("t-statistic:", t_statistic , "\n")
print("p-value:", p_value , "\n")

# Check if the p-value is less than 0.05
if p_value < 0.05:
    print("The Sample_1 Mean is significantly different from the Sample_2 Mean.")
else:
    print("The Sample_1 Mean is approximately equal to the Sample_2 Mean.")
```

### **Paired-Sample T-test (independent)**

### `H_0:` The Means of `Sample_1`, `Sample_2` and `Sample_3` are significantly equal.
### `H_1:` The Means of `Sample_1`, `Sample_2` and `Sample_3` are significantly different.

```python
sample_size = 500

# Take a random sample
sample_1 = df['Kilometers_Driven'].sample(n=sample_size, random_state=42)
sample_2 = df['Kilometers_Driven'].sample(n=sample_size, random_state=42)
sample_3 = df['Kilometers_Driven'].sample(n=sample_size, random_state=42)

# Perform the three-sample t-test
t_statistic, p_value = stats.f_oneway(sample_1, sample_2, sample_3)

# Print the results
print("Three-sample t-test results:\n")
print("t-statistic:", t_statistic , "\n")
print("p-value:", p_value , "\n")

# Check if the p-value is less than 0.05
if p_value < 0.05:
    print("The Means of the 'Sample_1', 'sample_2' and 'sample_3' are significantly different.")
else:
    print("The Means of the 'Sample_1', 'sample_2' and 'sample_3' are significantly equal.")
```

### **`Test for Homogeniety`**

### `H_0:` Variances of both samples are significantly equal.

### `H_1:` Variances of both samples are significantly unequal.

### 1. **Levene's Test** - Compare the variances of `Kilometers_Driven` and `Mileage` columns.

```python
# Perform Levene's Test
w_stat, p_val = stats.levene(df['Kilometers_Driven'], df['Mileage'])

print("w-statistic:", w_stat, '\n')
print("p-value:", p_val, '\n')

# print the results based on if else conditions
if p_val > 0.05:
    print('Variances of the "Kilomerters_Driven" and "Mileage" are significantly equal (fail to reject H0)')
else:
    print('Variances of the "Kilomerters_Driven" and "Mileage" are significantly unequal (reject H0)')
```

### Bartlett's Test - Compare the variances of `Kilometers_Driven` and `Price_lac` columns.

```python
# Perform Bartlett's Test
t_stat, p_val = stats.bartlett(df['Kilometers_Driven'], df['Price_lac'])

print("t-statistic:", t_stat , '\n')
print("p-value:", p_val , '\n')

# print the results using if else conditions
if p_val > 0.05:
    print('Variances of the "Kilomerters_Driven" and "Price_lac" are significantly equal (fail to reject H0)')
else:
    print('Variances of the "Kilomerters_Driven" and "Price_lac" are significantly unequal (reject H0)')
```

### `Hypothesis-03:` - Z-test (two tail)
### `H_0:` The `Price` means of `Diesel and Petrol Vehicles` are significantly equal.
### `H_1:` The `Price` means of `Diesel and Petrol Vehicles` are significantly unequal.

```python
diesel_prices = df[df['Fuel_Type'] == 'Diesel']['Price_lac']
petrol_prices = df[df['Fuel_Type'] == 'Petrol']['Price_lac']

# Perform the Shapiro-Wilk test to check the normality of the data.
statistic, p = stats.shapiro(diesel_prices)

# Print the results using if-else condition.
print('p-value =', p , '\n')
if p > 0.05:
    print('The "Diesel Cars" prices are Normally Distributed.\n')
else:
    print('The "Diesel Cars" prices are not Normally Distibuted.\n')

print("----------------------------------------")

# Perform the Shapiro-Wilk test to check the normality of the data.
statistic, p = stats.shapiro(petrol_prices)

# Print the results using if-else condition.
print('p-value =', p , '\n')
if p > 0.05:
    print('The "Petrol Cars" prices are Normally Distributed.\n')
else:
    print('The "Petrol Cars" prices are not Normally Distibuted.\n')
```

**Applying two-tail `Z-test`**

```python
# Calculate sample means and standard deviations
mean_diesel = np.mean(diesel_prices)
std_diesel = np.std(diesel_prices, ddof=1)
n_diesel = len(diesel_prices)

mean_petrol = np.mean(petrol_prices)
std_petrol = np.std(petrol_prices, ddof=1)
n_petrol = len(petrol_prices)

# Perform z-test
z_value = (mean_diesel - mean_petrol) / np.sqrt((std_diesel**2 / n_diesel) + (std_petrol**2 / n_petrol))

# Set significance level (alpha)
alpha = 0.05

# Calculate critical z-value for a two-tailed test
critical_z_value = stats.norm.ppf(1 - alpha/2)

# Compare z-value to critical z-value
if np.abs(z_value) > critical_z_value:
    print("There is a significant difference in mean prices. (reject H_0)")
else:
    print("There is no significant difference in mean prices. (fail to reject H_0)")
```

The report is structured to address the key questions posed, outline findings, discuss limitations, offer future recommendations, and conclude with a summary of insights gained.

#### **Data Overview:**

- The dataset used for this analysis contains a rich array of attributes related to use cars, including information on the city, car type (diesel or petrol), seating capacity, prices, fuel consumption, and more.

#### **Key Findings:**
1. City-Wise Car Sales by Fuel Type:

    + Our analysis indicates that diesel cars enjoy greater demand in most cities. However, exceptions exist in cities like `Kolkata, Kochi,  Pune and Mumbai`, where petrol cars dominate the market.

2. Seating Capacity Preferences:

    + City-wise demand analysis based on seating capacity reveals a pronounced preference for cars with a seating capacity of 5.

3. Price Trends by Fuel Type:

    + Petrol cars consistently demonstrate lower average prices in comparison to diesel cars.

4. Fuel Efficiency Analysis:

    + Diesel cars generally exhibit minor superiority in terms of mileage when compared to their petrol counterparts.

5. Sales Trends over Time:

    + The analysis suggests that there is a surge in the sale of diesel cars from `(2012 to 2017) `potentially due to their enhanced fuel efficiency.

6. Seating Capacity Insights:

    + On average, diesel cars have a higher seating capacity compared to petrol cars.

7. The average prices categorized by both location and fuel type:

    + When considering all cities, petrol cars are generally more affordable than diesel cars.

8. Price Comparison between Manual and Automatic Cars:

    + The general trend of average car prices exhibit an annual increase. Additionally, it is evident that the prices of Automatic cars consistently surpass those of Manual cars throughout the entire graph. Notably, in the years 1999, 2000, 2001, and 2002, there are no instances of Automatic cars available.

9. Sales Comparison between Manual and Automatic Cars:

    + Manual cars outsell automatic cars by a significant margin. The sales of manual cars show a substantial increase from 2010 to 2018 , It might be because of the low prices of the manual cars that can be seen in `Figure-12`. 

10. Mileage Comparison between Manual and Automatic Cars:

    + The manual cars generally have higher mileage than automatic cars. Greater presence of the manual cars in the data set can be attributed to this trend. 
11. Brand Analysis Based on Average Price:

    + The data unveils a clear distinction in pricing between different car brands. High-end brands such as Lamborghini, Bentley, and Porsche command top-tier prices, while Ambassador, Tata, Datsun, and Chevrolet offer more affordable options.

12. Brand Comparison on Fuel Efficiency:

    + Datsun emerges as a standout performer in terms of fuel efficiency (mileage), recording an impressive 21.59 kmpl. In contrast, Lamborghini lags behind with a meager mileage of 6.4 kmpl. Other brands fall within the range of 6.4 kmpl to 21.59 kmpl.
13. Insights from Data Correlations:

    + The analysis establishes a positive correlation between Power and Engine, while Mileage exhibits an inverse relationship with both Power and Engine. The correlation between Power and Engine is a robust 0.8, while the correlations between Mileage and Power (-0.5) and Mileage and Engine (-0.6) point to clear trends.
